```python
# The Mandelbrot set is a two-dimensional set with a relatively simple definition that exhibits great complexity, especially 

# as it is magnified. It is popular for its aesthetic appeal and fractal structures. The set is defined in the complex plane 

# as the complex numbers c for which the function fc(z) = z2 + c does not diverge to infinity when iterated starting at z = 0,

# i.e., for which the sequence fc(0), fc(fc(0)), etc., remains bounded in absolute value. The purpose of this notebook is to

# demonstrate and explain the calculation and visualization of the Mandelbrot set in python utilizing the numpy, sympy, math

# and matplotlib libraries.
```

```python
import numpy as np
import matplotlib.pyplot as plt

i_max = 256
height = 1600
width = 1600
xmin = -2
xmax = 1
ymin = -1.5
ymax = 1.5

def mandelbrot(c, i_max): # Defining the mandelbrot function
    z = 0
    for n in range(i_max):  # iteration loop
        if abs(z) > 2:
            return n
        z = z**2+c # fc(z) = z2 + c
    return i_max

def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, i_max):
    r1 = np.linspace(xmin, xmax, width) # array for real part coordinates
    r2 = np.linspace(ymin, ymax, height) # array for imaginary part coordinates
    n3 = np.empty((width, height)) # empty array for iteration count
    
    for i in range(width): # loop through each column
        for j in range(height): # loop through each row
            c = r1[i] + 1j * r2[j] # calculate complex number for respective pixel
            n3[i, j] = mandelbrot(c, i_max) #store iteration count
    return r1,r2,n3
    
r1, r2, n3= mandelbrot_set(xmin, xmax, ymin, ymax, width, height, i_max)

plt.figure(figsize=(10, 10))
plt.imshow(n3.T, extent=[xmin, xmax, ymin, ymax], cmap='prism')
plt.title('Mandelbrot Set')
plt.show()
```
